```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('cat.jpg', cv2.IMREAD_COLOR)

plt.imshow(image) ## opencv에서는 BGR 형식으로 되어있음
plt.show()

rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) ## BGR -> RGB
plt.imshow(rgb_image)
plt.show()

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) ## BGR -> GRAY

plt.imshow(gray_image)
plt.show()
plt.imshow(cv2.cvtColor(gray_image, cv2.COLOR_GRAY2RGB))
plt.show()
```

```python
print(image.shape) ## image shape 확인 (세로, 가로, RGB)
print(image.size)
```

```python
px = image[0:250,200:500] ## 이미지 객체의 특정 픽셀을 가르킴
plt.imshow(cv2.cvtColor(px, cv2.COLOR_BGR2RGB))
plt.show()
# print(px)
```

```python
import time

## 특정 픽셀 이미지 변환
""" 변환 방식 (1) """
start_time = time.time()
temp = image.copy()
for i in range(0, 100):
    for j in range(0, 100):
        temp[i, j] = [0, 0, 0]
print("--- %s seconds ---" % (time.time() - start_time))
plt.imshow(cv2.cvtColor(temp, cv2.COLOR_BGR2RGB))
plt.show()

""" 변환 방식 (2) """
start_time = time.time()
temp = image.copy()
temp[0:100, 0:100] = [0, 0, 0]
print("--- %s seconds ---" % (time.time() - start_time))
plt.imshow(cv2.cvtColor(temp, cv2.COLOR_BGR2RGB))
plt.show()

""" 변환 방식 (3) -- 붙여넣기 """
temp = image.copy()
# Numpy Slicing: ROI 처리 가능
roi = temp[110:140, 300:330]
# ROI 단위로 이미지 복사하기
temp[70:100, 340:370] = roi
plt.imshow(cv2.cvtColor(temp, cv2.COLOR_BGR2RGB))
plt.show()

""" 변환반식 (4) -- 색 변환 """
# temp[:, :, 2] = 0 ## BGR 에서 Index가 2인(python은 0부터 시작) Red에 대해 모두 0으로 변환
# plt.imshow(cv2.cvtColor(temp, cv2.COLOR_BGR2RGB))
# plt.show()
```
