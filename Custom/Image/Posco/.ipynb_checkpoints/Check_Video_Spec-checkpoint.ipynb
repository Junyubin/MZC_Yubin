## Chaning Foremat AVI to MP4

```python
import imageio
import os

def convertVideoToGifFile(inputFile, outputFile=None):	
    if not outputFile:
        outputFile = os.path.splitext(inputFile)[0] + ".gif"
		
    print("Converting {0} to {1}".format(inputFile, outputFile))

    reader = imageio.get_reader(inputFile)
    fps = reader.get_meta_data()['fps']

    writer = imageio.get_writer(outputFile, fps=fps)
	
    for i,im in enumerate(reader):
        writer.append_data(im)

    writer.close()
	
    print("\r\nConversion done.")

#Convert Input Files
convertVideoToGifFile("output_video/output_video.avi", "output_video/output_video.mp4")
```

```python
cap = cv2.VideoCapture('input_video/input_video.mp4')
inversed = []
frame_list = []
edge = []
# ë¹„ë””ì˜¤ ë§¤ í”„ë ˆìž„ ì²˜ë¦¬
while True: # ë¬´í•œ ë£¨í”„
    ret, frame = cap.read() # ë‘ ê°œì˜ ê°’ì„ ë°˜í™˜í•˜ë¯€ë¡œ ë‘ ë³€ìˆ˜ ì§€ì •
    if not ret: # ìƒˆë¡œìš´ í”„ë ˆìž„ì„ ëª»ë°›ì•„ ì™”ì„ ë•Œ braek
        break            
    frame_list.append(frame)
    edge.append(cv2.Canny(frame, 60, 360))  # ì •ì§€í™”ë©´ì—ì„œ ìœ¤ê³½ì„ ì„ ì¶”ì¶œ
    inversed.append(~frame)  # ë°˜ì „

    # 30ms ê¸°ë‹¤ë¦¬ê³  ë‹¤ìŒ í”„ë ˆìž„ìœ¼ë¡œ ì „í™˜, Escëˆ„ë¥´ë©´ while ê°•ì œ ì¢…ë£Œ
    if cv2.waitKey(10) == 27:
        break

cap.release() # ì‚¬ìš©í•œ ìžì› í•´ì œ
cv2.destroyAllWindows()
```
