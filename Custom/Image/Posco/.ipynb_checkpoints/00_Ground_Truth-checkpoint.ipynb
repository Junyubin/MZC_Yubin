# Ground Truth
### 레이블 지정 작업 인력 설정
* AWS 콘솔 -> SageMaker -> Ground Truth -> 레이블링 인력 -> 프라이빗 -> 프라이빗 팀 만들기
* 팀 이름 입력 -> 프라이빗 팀 만들기
* 작업자 -> 새 작업자 초대 -> 작업자 이메일 주소 추가

## 팀 내부 작업자 설정
* Ground Truth -> 레이블링 인력 -> 프라이빗 -> 팀이름 선택 -> 작업자 -> 팀에 작업자 추가  
  



### 작업 세부 정보 지정
* AWS 콘솔 -> SageMaker -> Ground Truth -> 레이블링 작업 -> 레이블 지정 작업 생성
* 작업 이름은 위의 json 파일의 ground_truth_job_name에 해당되는 **gt-version-0**을 입력해줍니다.
* 입력 데이터 세트의 S3 위치는 아래의 출력물을 입력해줍니다.
* 출력 데이터 세트의 S3 위치 -> 새 위치 지정 후 아래의 출력물을 입력해줍니다.

````python
print('입력 데이터 세트의 S3 위치 : s3://{bucket}/{version}/images/'.format(bucket = input_json['s3_bucket'], version = input_json['job_id']))
print('출력 데이터 세트의 S3 위치 : s3://{bucket}/{version}/{results}/'.format(bucket = input_json['s3_bucket'], version = input_json['job_id'], results = input_json['yolo_output_dir']))
````
