## Data Load 및 흑백 변환


```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('/root/MZC_Yubin/Sorin/dataset_220524/B_Swoot/Sheet_0001_Camera_03_First_Top_Grab.bmp', cv2.IMREAD_GRAYSCALE)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)


plt.imshow(image)
plt.show()
```

## 이미지 히스토그램 시각화 

```python
# 그레이스케일 [0], BG [0,1], BGR [0,1,2]
hist = cv2.calcHist([image], [0], None, [256], [0,256]) # 꼭 리스트 형태로 입력 (image, channels, mask[분석영역, 전체이면 None], histSize[BINS값 256], range(0,256))

plt.plot(hist)
plt.show()
```

## 이미지 이진화

```python
import numpy as np
image = cv2.imread('/root/MZC_Yubin/Sorin/dataset_220524/B_Swoot/Sheet_0001_Camera_03_First_Top_Grab.bmp', cv2.IMREAD_GRAYSCALE)

thresh = 100 ## 임계값
maxValue = 255 ## 적용할 value
# th, dst = cv2.threshold(image, thresh, maxValue, cv2.THRESH_BINARY)
th, dst = cv2.threshold(image, thresh, maxValue, cv2.THRESH_TRUNC)


plt.figure(figsize=(10,5))
plt.subplot(121),plt.imshow(image, cmap='gray'),plt.title('original image',fontsize=15)
plt.xticks([]), plt.yticks([])

plt.subplot(122),plt.imshow(dst,cmap='gray'),plt.title('filtered image',fontsize=15)
plt.xticks([]), plt.yticks([])

plt.show()
```

## 외곽선 검출

```python
import cv2
import numpy as np

# img = cv2.imread('/root/MZC_Yubin/Sorin/dataset_220524/B_Swoot/Sheet_0001_Camera_03_First_Top_Grab.bmp')                                # 이미지 불러오기
# imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)              # 이미지 그레이 전환

img = cv2.cvtColor(dst, cv2.COLOR_BGR2GRAY)
imgray = cv2.cvtColor(dst, cv2.COLOR_BGR2GRAY)

ret, thresh = cv2.threshold(imgray, 120, 255, 0)                                         # 흑과 백으로 임계(threshold) 분할
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)   # contour(외곽선)를 찾아냄.(연속된 좌표점)

cv2.drawContours(img, contours, -1, (0, 255, 0), 2)     # contour(외곽선)을 그림, 초록색(0 255 0), 두께 2로
cnt = contours[0]                                       # cnt변수에 Contour[0]에 있는 2차원 연속된 좌표를 넣음.

# 윤곽(convex)정보 휙득
hull = cv2.convexHull(cnt, returnPoints=False)

defects = cv2.convexityDefects(cnt, hull)   #

# for i in range(defects.shape[0]):

#     s, e, f, d = defects[i, 0]
#     start = tuple(cnt[s][0])
#     end = tuple(cnt[e][0])
#     far = tuple(cnt[f][0])

#     # 바깥 최 외곽선을 이은 line 표시 (파랑색)
#     cv2.line(img, start, end, [255, 0, 0], 2)

#     # 내부의 꼭지점에 좌표점 표시 (빨간색)
#     cv2.circle(img, far, 5, [0, 0, 255], -1)



plt.imshow(img)
plt.show()
```
