### 2. 데이터 확인

```python
%load_ext autoreload
%autoreload 2

import fastai
from fastai.vision import *
# defaults.device = torch.device('cuda') # gpu를 사용하는 경우만

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

import warnings
warnings.filterwarnings('ignore')

print(fastai.__version__)
```

```python
path = Path('PoC_data')
path.ls()
```

```python
data = ImageDataBunch.from_folder(
    path, 
    train='train',
    valid='val',
    ds_tfms=get_transforms(do_flip=False),  # 데이터셋 변형 함수
    size=224,  # 이미지 사이즈
    bs=16,     # 배치사이즈
    num_workers=8 # 워커 개수
)
```

```python
# 클래스 명칭과 클래스 개수를 출력
print(data.classes, data.c)

# train과 valid 데이터 개수 확인
print(len(data.train_ds), len(data.valid_ds))
```

```python
# 위 get_transforms() 함수로 변형된 이후의 이미지를 출력
data.show_batch()
```

### 3. ResNet34 훈련

딥러닝 모델의 최적화에 대한 이해를 도울 수 있는 자료입니다. 이번에 사용된 `fit_one_cycle`에 대한 자세한 설명은 뒷부분에 있습니다.
[https://iconof.com/1cycle-learning-rate-policy/](https://iconof.com/1cycle-learning-rate-policy/)

```python
learn = cnn_learner(data, models.resnet34, metrics = [accuracy])
```

```python
learn.fit_one_cycle(10)
```

```python
model_file_name = 'hsdm'
# 모델 저장
learn.save(model_file_name)
# 모델 로드
learn.load(model_file_name)
```

### 4. 모델 결과 분석

결과분석을 통해 어떤 클래스가 오류가 많은지 알 수 있습니다.

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix(figsize=(12,12), dpi=60)
```

```python
y_pred, y_true = learn.get_preds()
print(classification_report(y_true, np.argmax(y_pred, axis=1)))
```

```python
interp.most_confused()
```

```python
interp.plot_top_losses(k=12)
```
